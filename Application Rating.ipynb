```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"D:\GoIT\HW Python\applications(2.0).csv")
df.head()
```

Data cleaning

```python
df = df.drop_duplicates(subset='applicant_id')
```

```python
df['External Rating'] = df['External Rating'].fillna(0)
```

```python
df['Education level'] = df['Education level'].fillna('Середня')
```

Adding industry data

```python
industries = pd.read_csv(r"D:\GoIT\HW Python\industries(2.0).csv")
industries.head()
```

```python
df = pd.merge(df, industries)
df
```

Rating

```python
# working with time series
df['Applied at'] = pd.to_datetime(df['Applied at'], format='mixed', dayfirst=True, errors='coerce')
df['Applied on day'] = df['Applied at'].dt.day_name()
df
```

```python
# If the applicant's age is between 35 and 55, 20 points are added to the rating
df['Age'] = df['Age'].astype(int)
age = ((df['Age'] >= 35) & (df['Age'] <= 55)) * 20

# If the application was submitted not on a weekend, 20 points are added to the rating.
day = ~df['Applied at'].isin(['Saturday', 'Sunday']) * 20

# If the applicant is married, 20 points are added to the rating.
marital = (df['Marital status'] == 'Married') * 20

# If the applicant is located in Kyiv or the region, 10 points are added to the rating.
location = (df['Location'] == 'Київ чи область') * 10

# The 'Score' value from the industries.csv table is also added to the application (and ranges from 0 to 20 points)
industry = df['Score']

# If 'External Rating' is greater than or equal to 7, 20 points are added to the rating
df['External Rating'] = df['External Rating'].astype(int)
positive_rating = (df['External Rating'] >= 7) * 20

# If 'External Rating' is less than or equal to 2, 20 points are deducted from the rating
negative_rating = (df['External Rating'] <= 2) * -20

# Calculate the application rating according to the following conditions:

# Rating must be a number from 0 to 100
# Rating is the sum of the application's ratings across 6 criteria.
# Rating is zero if 'Amount' is missing or if 'External Rating' is zero.

df['Final Score'] = age + day + marital + location + industry + positive_rating + negative_rating
df['Final Score'] = df['Final Score'].clip(0, 100)
df['Final Score'] = df['Final Score'] * ~(df['External Rating'] == 0) * ~df['Amount'].isna()
df
```

In the resulting table, leave only applications with a rating greater than zero; these applications will be considered accepted.

```python
df = df[df['Final Score'] > 0]
df
```

Group the data from the resulting table by the week of application submission, and display the average rating of accepted applications in each week.

```python
df = df.set_index('Applied at')  # Set 'Applied at' as the index

weekly_avg_rating = df.resample('W')['Final Score'].mean()  # Resample by week and calculate mean

print(weekly_avg_rating)
```
